```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.datasets as dsets
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
import os

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

```python
def imshow(img):
    img = img / 2 + 0.5
    npimg = img.numpy()
    plt.figure(figsize=(10, 10))
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()

def show_images(dataloader):
    dataiter = iter(dataloader)
    images, _ = next(dataiter)

    imshow(torchvision.utils.make_grid(images))

show_images(dataloader)
```

```python
class Generator(nn.Module):
  def __init__(self):
    super(Generator, self).__init__()
    self.main = nn.Sequential(
        nn.ConvTranspose2d(100, 512, 4, 1, 0, bias=False),
        nn.BatchNorm2d(512),
        nn.ReLU(True),
        nn.ConvTranspose2d(512, 256, 4, 2, 1, bias=False),
        nn.BatchNorm2d(256),
        nn.ReLU(True),
        nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),
        nn.BatchNorm2d(128),
        nn.ReLU(True),
        nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),
        nn.BatchNorm2d(64),
        nn.ReLU(True),
        nn.ConvTranspose2d(64, 3, 4, 2, 1, bias=False),
        nn.Tanh()
    )

  def forward(self, input):
    return self.main(input)

class Discriminator(nn.Module):
  def __init__(self):
    super(Discriminator, self).__init__()
    self.main = nn.Sequential(
        nn.Conv2d(3, 64, 4, 2, 1, bias=False),
        nn.LeakyReLU(0.2, inplace=True),
        nn.Conv2d(64, 128, 4, 2, 1, bias=False),
        nn.BatchNorm2d(128),
        nn.LeakyReLU(0.2, inplace=True),
        nn.Conv2d(128, 256, 4, 2, 1, bias=False),
        nn.BatchNorm2d(256),
        nn.LeakyReLU(0.2, inplace=True),
        nn.Conv2d(256, 512, 4, 2, 1, bias=False),
        nn.BatchNorm2d(512),
        nn.LeakyReLU(0.2, inplace=True),
        nn.Conv2d(512, 1, 4, 1, 0, bias=False),
        nn.Sigmoid()
    )

  def forward(self, input):
    return self.main(input)
```

```python
Gen = Generator().to(device)
Disc = Discriminator().to(device)

loss_env = nn.BCELoss()

fixed_noise = torch.randn(64, 100, 1, 1, device=device)

optim_gen = optim.Adam(Gen.parameters(), lr=0.0002, betas=(0.5, 0.999))
optim_disc = optim.Adam(Disc.parameters(), lr=0.0002, betas=(0.5, 0.999))
```

```python
num_epos = 25
G_loss = []
D_loss = []

for epoch in range(num_epos):
  for i, data in enumerate(dataloader, 0):
    Disc.zero_grad()
    real_data = data[0].to(device)
    batch_size = real_data.size(0)
    labels = torch.full((batch_size,), 1, dtype=torch.float, device=device)
    output = Disc(real_data).view(-1)
    errD_real = loss_env(output, labels)
    errD_real.backward()

    noise = torch.randn(batch_size, 100, 1, 1, device=device)
    fake = Gen(noise)
    labels.fill_(0)
    output = Disc(fake.detach()).view(-1)
    errD_fake = loss_env(output, labels)
    errD_fake.backward()
    optim_disc.step()

    Gen.zero_grad()
    labels.fill_(1)
    output = Disc(fake).view(-1)
    errG = loss_env(output, labels)
    errG.backward()
    optim_gen.step()

    G_loss.append(errG.item())
    D_loss.append(errD_real.item() + errD_fake.item())

    if i % 100 == 0:
      print(f"Epoch: {epoch}/{num_epos}, Batch: {i}/{len(dataloader)}, D_loss: {errD_real.item() + errD_fake.item()}, G_loss: {errG.item()}")

plt.figure(figsize=(10, 5))
plt.title("Generator and Discriminator Loss")
plt.plot(G_loss, label="G")
plt.plot(D_loss, label="D")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
